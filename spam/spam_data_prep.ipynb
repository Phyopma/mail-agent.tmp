```python
import pandas as pd
import json
from sklearn.model_selection import train_test_split
```

```python
from datasets import load_dataset

ds = load_dataset("NotShrirang/email-spam-filter")
```

```python
df = pd.DataFrame(ds['train'])
```

```python
df = df[['text', 'label']]
```

```python
df.head()
```

```python
df['label'] = df['label'].replace({'ham': 'not spam', 'spam': 'spam'})
```

```python
df = df.rename(columns={'text': 'prompt', 'label': 'completion'})
```

```python
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)
valid_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)
```

```python
def save_jsonl(df, file_path):
    """Save dataframe as JSONL file."""
    with open(file_path, 'w') as f:
        for _, row in df.iterrows():
            json.dump(row.to_dict(), f)
            f.write('\n')
```

```python
save_jsonl(train_df, '/content/train.jsonl')
save_jsonl(valid_df, '/content/valid.jsonl')
save_jsonl(test_df, '/content/test.jsonl')

print("✅ Datasets saved as JSONL files:")
print("- /content/train.jsonl")
print("- /content/valid.jsonl")
print("- /content/test.jsonl")
```

```python
print("Number of training examples:", len(train_df))
print("Number of validation examples:", len(valid_df))
print("Number of testing examples:", len(test_df))
```
